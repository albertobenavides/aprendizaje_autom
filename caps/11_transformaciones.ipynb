# Transformaciones

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/sonar.csv', header = None)
df
```

```python
x = df.iloc[:, 0]
```

```python
# Se muestra la columna 1
plt.figure()
plt.hist(x)
plt.show()
```

## Recíproca

$$y = 1 / x$$

## Inversa de la recíproca

$$x = 1 / y$$

```python
plt.figure()
plt.hist(1/x)
plt.show()
```

## Raíz cuadrada

$$y = \sqrt{x}$$

### Inversa

$$x = y^2$$

```python
plt.figure()
plt.hist(np.power(x, 1/2))
plt.show()
```

### Ascombe

$$y = \sqrt{x + 3/8}$$

### Inversa de Ascombe

$$x = y^2 - 3/8$$

```python
plt.figure()
plt.hist(np.power(x + 3/8, 1/2))
plt.show()
```

### Freeman-Tukey

$$y = \sqrt{x} + \sqrt{x + 1}$$

```python
plt.figure()
plt.hist(np.power(x, 1/2) + np.power(x + 1, 1/2))
plt.show()
```

## Arcoseno

$$y = \arcsin{\sqrt{x}}$$

```python
plt.figure()
plt.hist(np.arccos(np.power(x, 1/2)))
plt.show()
```

## Tukey

$$y = 
\begin{cases}
    \lambda > 0 & x^\lambda \\
    \lambda = 0 & \log{x} \\
    \lambda < 0 & -(x^\lambda) \\
\end{cases}$$

```python
for l in np.arange(-3, 3, 0.5):
    plt.figure()
    plt.title(f'$\lambda = {l}$')
    if l > 0:
        plt.hist(np.power(x, l))
    elif l == 0:
        plt.hist(np.log(x))
    else:
        plt.hist(-np.power(x, l))
    plt.show()
```

## Box-Cox

$$y = 
\begin{cases}
    \lambda > 0 & x^\lambda - 1 \\
    \lambda = 0 & \log{x} \\
\end{cases}$$

```python
for l in np.arange(-3, 3, 0.5):
    plt.figure()
    plt.title(f'$\lambda = {l}$')
    if l > 0:
        plt.hist((np.power(x, l) - 1) / l)
    elif l == 0:
        plt.hist(np.log(x))
    plt.show()
```
